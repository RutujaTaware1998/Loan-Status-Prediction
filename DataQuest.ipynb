```python
X_train = train.drop(['Loan ID' , 'Purpose' ,'Loan Status','Rent' , 'renewable_energy'] ,axis =1)
```

```python
Y_train = train.loc[: ,'Loan Status']
```

```python
x = X_train.values
y = Y_train.values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_val = sc.transform(x_val)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
classifier = Sequential()
classifier.add(Dense(34, activation = 'relu', input_dim = 32, kernel_initializer = 'uniform'))
classifier.add(Dense(34, activation = 'relu',kernel_initializer = 'uniform'))
classifier.add(Dense(34, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
classifier.add(Dense(1, activation = 'sigmoid',kernel_initializer = 'uniform'))
adam = Adam(lr = 0.0001)
classifier.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
classifier.fit(x_train, y_train, validation_data = (x_val, y_val), batch_size = 50, epochs = 100)
```

```python
classifier = Sequential()
classifier.add(Dense(34, activation = 'relu', input_dim = 32, kernel_initializer = 'uniform'))
classifier.add(Dense(34, activation = 'relu',kernel_initializer = 'uniform'))
classifier.add(Dense(34, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
#classifier.add(Dense(15, activation = 'relu',kernel_initializer = 'uniform'))
classifier.add(Dense(1, activation = 'sigmoid',kernel_initializer = 'uniform'))
adam = Adam(lr = 0.0001)
classifier.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
```
